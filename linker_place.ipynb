```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.image as mpimg
```

```python
fig, axs = plt.subplots(1,3, figsize=(14,7))

img_sumSigma = mpimg.imread('Results_eps_attr10_eps_rep1_range2_attr_with_repTrue_n_vesicles1_sumSigma/image.png')
axs[0].imshow(img_sumSigma)
axs[0].set_title("sum of sigmas = 5.25")

img_vesSigmaLJ = mpimg.imread('Results_eps_attr10_eps_rep1_range2_attr_with_repTrue_n_vesicles1_vesicleSigmatimesLJ/image.png')
axs[1].imshow(img_vesSigmaLJ)
axs[1].set_title("sigma of vesicle*lj = 5.61")

img_vesSigmaLJ = mpimg.imread('Results_eps_attr10_eps_rep1_range2_attr_with_repTrue_n_vesicles1_sumSigmatimesLJ/image.png')
axs[2].imshow(img_vesSigmaLJ)
axs[2].set_title("sum of vesicles*lj = 5.89")
```

```python
data_sumSigma = pd.read_csv("Results_eps_attr10_eps_rep1_range2_attr_with_repTrue_n_vesicles1_sumSigma/data.csv", delimiter=r"\s+")
data_vesLJ = pd.read_csv("Results_eps_attr10_eps_rep1_range2_attr_with_repTrue_n_vesicles1_vesicleSigmatimesLJ/data.csv", delimiter=r"\s+")
data_sumLJ = pd.read_csv("Results_eps_attr10_eps_rep1_range2_attr_with_repTrue_n_vesicles1_sumSigmatimesLJ/data.csv", delimiter=r"\s+")
```

```python
fig, axs = plt.subplots(1,3, figsize=(14,7))
xs=data_sumLJ.index

axs[0].plot(data_sumSigma["TotEng"])
axs[0].plot(xs, [data_sumSigma["TotEng"].mean() for i in range(1001)])
axs[0].set_title("sum of sigmas = 5.25")

axs[1].plot(data_vesLJ["TotEng"])
axs[1].plot(xs, [data_vesLJ["TotEng"].mean() for i in range(1001)])
axs[1].set_title("sigma of vesicle*lj = 5.61")

axs[2].plot(data_sumLJ["TotEng"])
axs[2].plot(xs, [data_sumLJ["TotEng"].mean() for i in range(1001)])
axs[2].set_title("sum of vesicles*lj = 5.89")
```
